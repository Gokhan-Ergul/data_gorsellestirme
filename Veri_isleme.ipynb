## Boxplot
veri setinin hikayesi
* total_bill = vergiler dahil fatura
* tip = bahsis
* sex = ucret odeyenlerin cinsiyeti
* smoker = grupta sigara icen var mi (yes, no) (male = 0, female = 1)
* day = gun (3 = thur, 4 = fri, 5 = Sat, 6 = Sun)
* time = ne zaman? (0 = Day, 1 = Night)
* size = gruplar kacar kisiden olusuyor


```python
tips = sns.load_dataset('tips')
df = tips.copy()
df.head()
```

```python
# hangi gun daha cok para kazanildigini bulmak istiyorum.
df.groupby("day", observed=False)[['total_bill']].mean()
# pazar gunleri daha cok is yapildigini goruyorum...  peki aksam yemekleri mi yoksa oyle yemeklerinden mi daha cok
#gelir elde ediliyor?
df.groupby(['day', 'time'],observed=False)[['total_bill']].mean()
```

```python
#kadinlarin mi daha cok sigara iciyor erkekler mi?
df.groupby(['smoker', 'sex'],observed = False)[['smoker']].count()
```

```python
a = df.describe().T
a.iloc[0:,1:8]
```

```python
#sigara icenlerin bahsislere yansimasi nasil olur?
df.groupby(['smoker'],observed = False)[['tip']].mean() 
#sigara icen  daha cok bahsis birakmis
```

## Cizgi grafigi

```python
fmri = sns.load_dataset('fmri')
df = fmri.copy()
df.head()
df
```

fmri, beyine baglanan cihazdan alinan verilerin tablosu
* subject = verilerin toplandigi kisiler
* timepoint = zaman noktalari
* event = birbirinden farkli olaylari ifade ediyor stim(uyarmak)
* region = sinyalin toplandigi bolge
* signal = alinan sinyaller

```python
df.groupby('timepoint')[['signal']].count() #her bir zaman noktasinda esit signal toplanmis
```

```python
df.groupby('signal')[['signal']].count()
```

```python
df.shape # 1064 tane signal varmis dememek ki her veri birbirinden farkli
```

```python
df.groupby('timepoint')['signal'].describe()
```
